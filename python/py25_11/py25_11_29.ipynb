```python
import numpy as np
from numpy import pi
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
```

```python
def func():
    x = np.arange(-pi,pi,0.01)
    c = np.cos(x)
    s = np.sin(x)
    plt.figure(figsize = (3 * pi,3))
    plt.plot(x,c)
    plt.plot(x,s)
    ax = plt.gca()
    ax.spines[['top','right']].set_visible(False)
    ax.spines[['bottom','left']].set_position('zero')
    ax.set_xticks([-pi,-pi/2,0,pi/2,pi])
    ax.set_xticklabels(['-π','-π/2','0','π/2','π'])
func()
```

选取范围找水仙花数

```python
def func(start,end):
    for num in range(start,end):
        string = str(num)
        num_len = len(string)
        temp = 0
        for i in string:
            temp += (int(i) ** num_len)
        if num == temp:
            print(num)
func(10,1000)
```

```python
def func(start,end):
    li = []
    for num in range(start,end):
        flag = True
        for i in range(2,int(pow(num,0.5)) + 1):
            if num % i == 0:
                flag = False
                break
        if flag:
            li.append(num)
    print(li)
func(1,100)
```

```python
def func(num):
    li = []
    for i in range(num):
        li.append([])
        for j in range(num):
            li[i].append(0)
    for i in range(num):
        li[i][0] = 1
        li[i][i] = 1
    for i in range(2,num):
        for j in range(1,num):
            li[i][j] = li[i - 1][j] + li[i -1][j -1]
    for i in range(num):
        for j in range(num):
            if li[i][j] != 0:
                print(li[i][j],end = ' ')
        print()
func(10)
```
