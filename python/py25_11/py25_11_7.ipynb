```python
# 判断1000以内的完数
def funa():
    for i in range(1,1000):
        sum=0
        for j in range(1,int(i/2)+1):
            if i%j==0:
                sum+=j
        if sum==i:
            print(f'{i}是完数')
funa()
```

```python
# 找质数因子
def funa():
    num=int(input("输入一个自然数:"))
    print(num)
    a=2
    li=[]
    while num>=a:
        if num%a==0:
            li.append(a)
            num/=a
        else:
            a+=1
    for i in li:
        print(i,end=' ')
funa()
```

```python
# 找到列表中重复的数
def funa(li):
    deldata=[]
    for i in li:
        if li.count(i)>=2 and i not in deldata:
            deldata.append(i)
    return deldata
li=[1,2,3,3,7,7,9,12,13,12,45,23,1]
funa(li)
```
