numpy  
用于科学计算和数据计算

```python
import numpy as np
```

n维数组(ndarray)

```python
arr = np.array([1,2,3,4])
print(arr,type(arr))
arr = np.array([
    [1,2,3],
    [4,5,6]
])
print(arr)
```

ndarray中的属性  
shape:结构  
size:元素个数  

```python
arr = np.array([
    [1,2,3],
    [4,5,6]
])
print(arr.size)
print(arr.shape)
```

zeros,ones  
初始化数组  
zeros_like,ones_like  
将有值数组初始化，结构不变  
full(n,x)  
n为结构，填元组，x为充满的值

```python
arr = np.zeros((2,3))
print(arr)
arr = np.ones((3,4))
print(arr)
arr = np.array([
    [1,2,3],
    [4,5,6]
])
print(np.ones_like(arr))
print(np.full((3,4),4))
```

linspace  
linspace(0,10,num = 100)   
0~10之间取100个等差的数  
num默认为50  
endpoint = True默认，表示右边范围为包含]的

```python
arr = np.linspace(0,10,num = 10,endpoint = False)
print(arr)
```

fromfunction  


```python

arr = np.fromfunction(lambda x:x + 1,(10,))
print(arr)
print('\n')
arr = np.fromfunction(lambda x,y:(x + 1) * (y + 1),(9,9))
print(arr)
```

ndarray可以指定数据类型  
传参dtype

```python
arr = np.array([
    [1,2,3],
    [4,5,6],
],dtype = np.float64
)
print(arr)
```

np.sctypeDict  
查询numpy的数据类型

randint(n,m,(x,y))  
n~m结构为(x,y)的整数数组  
randn(x,y,z)  
结构为(x,y,z)的符合正态分布的数组  
uniform(n,m,(x,y))  
n~m结构为(x,y)的数组  

```python
arr = np.random.randint(0,150,(2,3))
print(arr)
arr = np.random.randn(2,2,2)
print(arr)
arr = np.random.uniform(0,150,(2,3))
print(arr)
```

数组的计算

```python
arr1 = np.array([
    [1,2,3],
    [4,5,6]
])
print(arr1 * 2 - 1)
arr2 = np.array([
    [6,5,4],
    [3,2,1]
])
np.add(arr1,arr2)
print(arr1 * arr2)
```

矩阵运算  
np.dot

```python
a1 = np.random.randint(0,100,(2,3))
a2 = np.random.randint(0,100,(3,2))
display(a1,a2)
np.dot(a1,a2)
```

np,arange()  
生成等差的数，可以为小数

```python
np.arange(0,10,0.5)
```

索引

```python
arr = np.array([
    [1,2,3],
    [4,5,6]
])
print(arr[0][0])
print(arr[0][0:2])
print(arr[:2,:2])
```

bool索引

```python
arr = np.array([
    [1,2,3],
    [1,2,1]
])
arr[arr == 1] = 5
print(arr)
arr = np.random.randn(2,3)
arr[arr < 0] = 0
print(arr)
arr = np.array([
    [1,2,3],
    [1,2,1]
])
arr[(arr == 1) | (arr == 2)] = 5
print(arr)
```

np.sum()  
axis = 1 计算横轴
axis = 0 计算数轴

```python
arr = np.random.randint(1,10,(2,3))
np.sum(arr,axis = 1)
np.sum(arr,axis = 0)
```
