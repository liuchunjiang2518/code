```python
import math
def func(start,end):
    li = []
    for num in range(start,end + 1):
        flag = True
        for i in range(2,int(math.sqrt(num)) + 1):
            if num % i == 0:
                flag = False
                break
        if flag:
            li.append(num)
    return li
print(func(1,100))
```

```python
def func(start,end):
    li = []
    for i in range(start,end):
        for j in range(start,end):
            for k in range(start,end):
                if i != j != k and i != k:
                    res = str(i) + str(j) + str(k)
                    li.append(res)
    return li
print(func(1,5))
```

```python
def func():
    for num in range(1,1000):
        temp = 0
        for i in range(1,int(num / 2) + 1):
            if num % i == 0:
                temp += i
        if num == temp:
            print(f'{num}是完数')
func()
```

```python
def func(num):
    i = 2
    li = []
    while num >= i:
        if num % i == 0:
            li.append(i)
            num /= i
        else:
            i += 1
    return li
print(func(56))
```

```python
def func(num):
    temp = str(num)[::-1]
    if temp == str(num):
        print(f'{num}是回文数')
func(12321)
func(123)
```

```python
def func(li):
    temp = []
    for i in li:
        if li.count(i) > 1 and i not in temp:
            temp.append(i)
    return temp
print(func([1,2,3,2,3,4,5,6,4]))
```

```python
def func():
    li = []
    k = 1
    sum = 0
    for i in range(3):
        li.append([])
        for j in range(3):
            li[i].append(k)
            k += 1
            if i == j:
                sum += li[i][j]
    return sum
print(func())
```
