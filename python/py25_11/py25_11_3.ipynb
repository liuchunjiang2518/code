```python
# 判断是否为素数
def funa():
    a=int(input('输入整数'))
    flag=True
    if a<=1:
        print(f'{a}不是素数')
    if a==2:
        print(f'{a}是素数')
    for i in range(3,int(pow(a,0.5))+1,2):
        if a%i==0:
            flag=False
            break
    print(f'{a}是素数') if flag else print(f'{a}不是素数')
funa()
```

```python
# 求阶乘
def funa(a):
    if(a==1):
        return 1
    return funa(a-1)*a
funa(4)
```

```python
# 求水仙花数
def funa(num):
    n=len(str(num))
    result=0
    temp=num
    while(temp):
        digtal=temp%10
        result+=digtal**n
        temp=temp//10
    if result==num:
        return True
    else:
        return False
funa(1634)
```
