```python
# 元组：不可修改
# 空元组
tu=tuple()
tu=()
# 赋值
tu=(1,'Hello',True)
print(type(tu),tu)
```

```python
tu=(123)
print(type(tu))
# 单个元素加空逗号
tu=(123,)
```

```python
tu=(1,2,3),(4,5,(6,7,8))
print(type(tu),tu)
# 下标取元素
tu[1][2][0] # 对应6
```

```python
# 方法
# index
tu[0].index(2) # 1
# count
tu[1].count((6,7,8)) # 1
# len
len(tu[1]) # 3
```

```python
# 遍历
# while
idx=0
while idx < len(tu[1]):
    print(tu[1][idx])
    idx += 1
for i in tu[1]:
    print(i)
```

```python
# 元组不可修改，但元组内的列表可以修改
# tu[0]=(2,3,4)
tu = (1,2,[3,4,5])
tu[2][0] = 9
tu
```

```python
tu=('万春宏',11,['football','music'])
tu.index(11)
tu[0]
del tu[2][0]
tu[2].append('coding')
tu
```

```python
# 字符串
str_ = 'hello world'
val = str_[-6]
val
# 字符串无法修改
# 方法
str_.index('ll') # 起始下标
# replace
str_new = str_.replace('world','WanChunhong')
str_new
# split按指定字符分割
for i in str_.split('l'):
    print(i)
# strip 去除空格和回车或指定字符,一个字符一个字符执行，不按顺序
str_ = '123WanChunhong321'
print(str_.strip('123'))
str_.count('1')
num = len(str_)
num
```
