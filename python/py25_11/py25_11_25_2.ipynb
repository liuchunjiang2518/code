不定长参数  
*args:会被合并为一个元组  
**kwargs:会被合并成一个字典

```python
def test(*args):
    print(args)
test('万','春','宏')
def test(**kwargs):
    print(kwargs)
test(one = 1,two = 2,three = 3)
```

函数作为参数传递  
匿名函数

```python
def test(func,x,y):
    num = func(x,y)
    return num
def sum(x,y):
    return x + y
def mod(x,y):
    return x % y
print(test(mod,1,2))
test(lambda x,y:x - y,12,2)
```

文件的关闭  
close()关闭文件的占用

with open 语法操作文件  
可以在操作后自动close

```python
with open('D:/Code/code/test.txt','r',encoding = 'UTF-8') as f:
    for line in f:
        print(line)
```

读取文件word_for25_11_25_2.txt并统计itheima出现次数

```python
with open('word_for25_11_25_2.txt') as f:
    f_str = f.read()
    count = f_str.count('itheima')
print(count)

with open('word_for25_11_25_2.txt') as f:
    count = 0
    for line in f:
        count += line.count('itheima')
    print(count)

with open('word_for25_11_25_2.txt') as f:
    count = 0
    lines = f.readlines()
    for line in lines:
        count += line.count('itheima')
    print(count)
```
