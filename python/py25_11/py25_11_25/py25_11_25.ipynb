```python
# 字典
# 空字典
dic = {}
dic = dict()
dic = {'Wan':12,'Chun':18,'Hong':206}
# 不能有重复的key，保留后面的值
print(dic['Hong'])
# dict可以嵌套，key不能是dict
dic = {1:12,2:113,3:{1:23,2:98,3:32}}
print(dic[3][1])
```

```python
# 新增，更新元素
dic[2] = {1:19,2:78}
print(dic[2][2])
dic[4] = 'waaaagh'
print(dic)
```

```python
# 删除清空元素
num = dic.pop(1) # pop返回被删除的value
print(dic)
print(num)
dic.clear()
print(dic)
```

```python
dic = {'万':{'salary':3000,'d_name':'科技部','rank':1},
       '春':{'salary':5000,'d_name':'市场部','rank':2},
       '宏':{'salary':7000,'d_name':'市场部','rank':3},
       '是':{'salary':4000,'d_name':'科技部','rank':1},
       '人':{'salary':5000,'d_name':'市场部','rank':2}}
print(dic)
for name in dic:
    if dic[name]['rank'] == 1:
        dic[name]['rank'] += 1
        dic[name]['salary'] += 1000
print(dic)
```

```python
# 数据容器的常用通用函数
# len(),max(),min()
li = [1,2,9,5]
print(max(li))
dic = {'key1':99,'key2':12}
print(max(dic))
dic = dict(one = 1,two =2,three = 3)
print(dic)
# 将dict转换为list保留key
print(list(dic))
print(tuple(dic))
print(str(dic)) # str(dic) == '{'one': 1, 'two': 2, 'three': 3}'
print(set(dic))
```
