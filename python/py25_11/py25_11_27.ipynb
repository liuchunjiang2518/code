字典创建Series  
key->index  
value->values  
如果自己加入index，并且与字典的key不匹配  
Series会使用自己添加的index，并且无对应值的index为空(NaN)  
与index不对应的key不显示

```python
s = Series({'Beijing':1500,'Chendu':1200},index = ['Beijing','Shanghai'])
print(s)
s[pd.isnull(s)] = 0
print(s)
s[pd.notnull(s)] = 1000
print(s)
```

Series之间可以计算

```python
data1 = {'Beijing':1500,'Nanjing':1300,'Chendu':1200}
data2 = {'Beijing':2000,'Nanjing':1600,'Shanghai':1400}
s1 = Series(data1)
s2 = Series(data2)
print(s1 + s2)
```

分别计算每一单的总价

```python
df2 = df.groupby(by = ['order_id']).sum()
s = df2['item_price_float']
s.sum()
price_avg = (s.sum()) / (s.shape[0])
print(round(price_avg,2))
```

```python
round(df.groupby(['order_id']).sum()['item_price_float'].mean(),2)
```
