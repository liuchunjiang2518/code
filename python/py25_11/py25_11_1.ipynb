```python
def funb():
    x=np.arange(10)
    y=x**2
    plt.figure(figsize=(6,8)) # 84
    plt.plot(x,y,'-x')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    ax=plt.gca()
    for i,j in zip(x,y):
        ax.text(i,j,(i,j))
funb()
```

```python
# 
def funa():
    x=np.arange(50)
    colors=['r','b','y','g','m','c']
    line_kind={
        'circle':'-o',
        'start':'*',
        'traggle_down':'v',
        'dash_dot':'-',
        'dash_line':'--',
        'x-mark':'-x'
    }
    plt.figure(figsize=(12,8))
    for index,key in zip(np.arange(6),line_kind.keys()):
        y=np.random.randn(50)
        plt.plot(x,y,line_kind[key],color=colors[index],label=key)
funa()
```
