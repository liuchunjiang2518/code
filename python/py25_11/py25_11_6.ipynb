```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams["font.family"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
```

```python
def funa():
    x=np.arange(0,10)
    y=x**2
    plt.figure()
    plt.plot(x,y,'-x')
    plt.xlabel('x轴')
    plt.ylabel('y轴')
    ax=plt.gca()
    for i,j in zip(x,y):
        ax.text(i,j,(i,j))
funa()
```

```python
def funa():
    x=np.arange(100)
    colors=['r','y','g','b','c','m']
    kind={
        'circle':'o',
        'start':'*',
        'traggle_donw':'v',
        'dash_dot':'-',
        'dash_line':'--',
        'x_mark':'x'
        }
    plt.figure(figsize=(12,8))
    for index,key in zip(np.arange(6),kind.keys()):
        y=np.random.randn(100)
        plt.plot(x,y,kind[key],color=colors[index])
funa()
    
```

```python
from numpy import pi
def funa():
    x=np.linspace(-pi,pi,256)
    c=np.cos(x)
    s=np.sin(x)
    plt.figure(figsize=(3*pi,3))
    plt.plot(x,c)
    plt.plot(x,s)
    plt.xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$'])
    plt.yticks([-1,0,1])
    ax=plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_position(('data',0))
    ax.spines['left'].set_position(('data',0))
funa()
```

```python
# 柱状图
def funa():
    citys=['Beijing','Shanghai','Chendu','Nanjing']
    price=[150,230,300,250]
    plt.bar(citys,price,align='center',color='steelblue',alpha=0.8)
    plt.title('House_price')
funa()
```

```python
def funa():
    citys=['Beijing','Shanghai','Chendu','Nanjing']
    price=[150,230,300,250]
    plt.barh(citys,price,align='center',color='steelblue',height=0.5)
    plt.xlim([0,400])
    for x,y in zip(np.arange(4),price):
        plt.text(y,x,y)
funa()
```

```python
def funa():
    price2024=[123,245,236]
    price2025=[135,298,150]
    citys=['Beijing','Xi\'an','Chendu']
    bar_width=0.35
    plt.bar(np.arange(3),price2024,color='steelblue',alpha=0.8,width=bar_width)
    plt.bar(np.arange(3)+bar_width,price2025,color='indianred',alpha=0.8,width=bar_width)
    plt.xticks(np.arange(3)+bar_width/2,citys)
    plt.ylim(0,400)
    for x,y in zip(np.arange(3),price2024):
        plt.text(x-0.05,y,y)
    for x,y in zip(np.arange(3)+bar_width,price2025):
        plt.text(x-0.05,y,y)
funa()
```
