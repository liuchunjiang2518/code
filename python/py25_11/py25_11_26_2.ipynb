map  
用func处理可迭代对象  
map的应用  
函数：传入一个数字列表  
返回一个新列表，每个数加10

```python
def test_map(arr,func,x):
    li = []
    for i in arr:
        li.append(func(i,x))
    return li
sum = lambda x,y:x + y
arr = [1,2,3,4]
print(test_map(arr,sum,10))
```

```python
li = [1,2,3,4]
li_new = list(map(lambda x:x + 10,li))
print(li_new)
```

需求：  
将li = ['1:woshjc','2:wnjxs','1:opwqqw','3:abduwbca']引号及数字去掉  
使用map

```python
li = ['1:woshjc','2:wnjxs','1:opwqqw','3:abduwbca']
li_new = list(map(lambda x:x[2:],li))
print(li_new)
```

需求：  
li = ['1:woshjc','2:wnjxs','1:opwqqw','3:abduwbca']只保留含'1:'的字符串

```python
li = ['1:woshjc','2:wnjxs','1:opwqqw','3:abduwbca']
li_new = []
def test(str_):
    if '1:' in str_:
        li_new.append(str_)
list(map(test,li)) # 加入list()，强制执行map()
print(li_new)
```

filter  
用func过滤可迭代对象

```python
li = ['1:woshjc','2:wnjxs','1:opwqqw','3:abduwbca']
def test(func,arr):
    li = []
    for i in arr:
        if func(i):
            li.append(i)
    return li
v = test(lambda x:'1:' in x,li)
print(v)


v = filter(lambda x:'1:' in x,li) # lambda x:x.startswith('1:')
print(list(v))
```

zip  
合并一些内容  
合并多个可迭代对象  
合并后的元素为tuple  
返回对象是迭代器

```python
name = ['wan','chun','hong']
age = [12,31,99]
id_ = ['31','xx1','39']
v = zip(name,age,id_)
print(list(v))
```

reduce  
用func累计可迭代对象

```python
from functools import reduce
li = [1,2,3,4]
reduce(lambda x,y:x + y,li)
```
