list的遍历

```python
li=[1,2,3,4,5,6]
for i in li:
    print(i)
idx=0
while idx<len(li):
    print(li[idx])
    idx+=1
```

```python
def listTraverse(list):
    for i in list:
        print(i)
listTraverse(li)
```
