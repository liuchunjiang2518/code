文件的写出  
write写入  
flush刷新(将内容真正写入硬盘)  
close也有flush功能

'w'模式会清空原来的内容

```python
with open('test.txt','w') as f:
    f.write('Hello World')
    f.flush()
    f.write('WanChunhong')
    f.flush()
```

文件的综合练习  
读取文件bill.txt  
备份文件bill.txt.bak  
将标记为测试的内容去除


```python
with open('bill.txt','r',encoding = 'UTF-8') as f:
    f_read = f.read()
    print(f_read)
with open('bill.txt','r',encoding = 'UTF-8') as f:
    with open('bill.txt.bak','w') as b:
        for line in f:
            if line.count('测试') == 0:
                b.write(f'{line}\n')
            b.flush()


f = open('bill.txt','r')
f_read = f.readlines()
b = open('bill.txt.bak','w')
for line in f_read:
    # if line.count('测试') > 0:
    #     continue
    # b.write(f'{line}\n')
    if '测试' not in line:
        b.write(line)
```
